### Ejercicio 1 - Investigar Como?

#### Resolver las siguientes ecuaciones:


$$12x + 5y + z= 6$$
$$6x + 10y + z = 8$$
$$3x + 15y + z = 4$$



#### Utilizando un método diferente al enseñado  $A*x =B$

```python
import numpy as np

# Matriz aumentada del sistema
M = np.array([
    [12,  5,  1, 6],
    [6,  10,  1, 8],
    [3,  15,  1, 4]
], dtype=float)
```

```python
# Paso 1: hacer pivote en fila 1
M[1] = M[1] - (M[1][0] / M[0][0]) * M[0]
M[2] = M[2] - (M[2][0] / M[0][0]) * M[0]

# Paso 2: eliminar debajo del segundo pivote (posición [1][1])
M[2] = M[2] - (M[2][1] / M[1][1]) * M[1]

# Matriz triangular superior
print("Matriz triangular superior:")
print(M)
```

```python
z = M[2][3] / M[2][2]
y = (M[1][3] - M[1][2] * z) / M[1][1]
x = (M[0][3] - M[0][1] * y - M[0][2] * z) / M[0][0]

print(f"x = {x:.4f}, y = {y:.4f}, z = {z:.4f}")
```

### Ejercicio 2 - Investigar Como?

Integrar **Ecuaciones diferenciales ordinarias** usar la función `odeint` del paquete `integrate`, que permite integrar sistemas del tipo:

$$ \frac{d\mathbf{y}}{dt}=\mathbf{f}\left(\mathbf{y},t\right)$$

con condiciones iniciales $\mathbf{y}(\mathbf{0}) = \mathbf{y_0}$.

```python
# Pista utilizar.......
from scipy.integrate import odeint
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir la función f(y, t) = dy/dt
def f(y, t):
    return -3 * y + 2

# Condición inicial
y0 = 4

# Rango de tiempo (de 0 a 5 segundos, 100 puntos)
t = np.linspace(0, 5, 100)

# Resolver la EDO con odeint
solucion = odeint(f, y0, t)

# Graficar la solución
plt.plot(t, solucion)
plt.title("Solución de la EDO: dy/dt = -3y + 2")
plt.xlabel("Tiempo (t)")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()
```

<div class="well">Realizar un ajuste polinomico en numpy del archivo <code>covid.txt</code> (archivo adjunto)
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos del archivo
datos = np.loadtxt('covid.txt')

# Verificación
print(datos.shape)
print(datos)

# Si solo hay una columna
if datos.ndim == 1:
    casos = datos
    dias = np.arange(len(casos))
else:
    dias = datos[:, 0]
    casos = datos[:, 1]

# Crear un rango de días para graficar y ajustar
dias_suavizados = np.linspace(min(dias), max(dias), 300)

# Ajustes polinómicos
coef_grad1 = np.polyfit(dias, casos, 1)
coef_grad2 = np.polyfit(dias, casos, 2)
coef_grad3 = np.polyfit(dias, casos, 3)

poly_grad1 = np.poly1d(coef_grad1)
poly_grad2 = np.poly1d(coef_grad2)
poly_grad3 = np.poly1d(coef_grad3)

# Evaluar los polígonos
casos_grad1 = poly_grad1(dias_suavizados)
casos_grad2 = poly_grad2(dias_suavizados)
casos_grad3 = poly_grad3(dias_suavizados)

```

### Ejercicio 4 - Investigar Como?

<div class="well">Crear un programa que permita hacer operaciones con matrices en <strong>Python</strong> utilizando los módulos <strong>NumPy</strong> y <strong>SciPy</strong> para resolver la inversa, la multiplicación y la solución de un sistema de ecuaciones lineales.


</div>


```python
import numpy as np
from scipy import linalg

# ------------------------
# 1. Inversa de una matriz
# ------------------------
print("1. MATRIZ INVERSA\n")

A = np.array([[2, 1],
              [5, 3]])

print("Matriz A:\n", A)

try:
    A_inv = linalg.inv(A)
    print("Inversa de A:\n", A_inv)
except np.linalg.LinAlgError:
    print(" La matriz A no es invertible.")

# ------------------------
# 2. Multiplicación de matrices
# ------------------------
print("\n2. MULTIPLICACIÓN DE MATRICES\n")

B = np.array([[1, 2],
              [0, 1]])

C = np.dot(A, B)
print("Matriz B:\n", B)
print("A * B:\n", C)

# ------------------------
# 3. Sistema de ecuaciones lineales
# ------------------------
print("\n3. SOLUCIÓN DE SISTEMA DE ECUACIONES\n")

# Sistema: Ax = b
# 2x + y = 8
# 5x + 3y = 18

b = np.array([8, 18])
x = linalg.solve(A, b)

print("Vector b:", b)
print("Solución x del sistema Ax = b:", x)
```

### Ejercicio 5 - Investigar Como?

<div class="well">Crear un programa que permita resolver una <strong>interpolación utilizando polinomios de Lagrange</strong> en Python <strong> SciPy </strong> .


</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# -----------------------------
# Datos conocidos (x, y)
# -----------------------------
x = np.array([0, 1, 2, 3])
y = np.array([1, 3, 2, 5])

# -----------------------------
# Calcular el polinomio de Lagrange
# -----------------------------
P = lagrange(x, y)

print("Polinomio de Lagrange:")
print(P)

# -----------------------------
# Evaluar el polinomio en puntos finos
# -----------------------------
x_fino = np.linspace(min(x), max(x), 200)
y_interp = P(x_fino)

# -----------------------------
# Graficar
# -----------------------------
plt.plot(x, y, 'ro', label='Puntos originales')
plt.plot(x_fino, y_interp, 'b-', label='Interpolación de Lagrange')
plt.title("Interpolación con Polinomio de Lagrange")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### Ejercicio 6 - Investigar Como?

<div class="well"><strong> Graficar las siguientes funciones:</strong>
    
1. $f(x) = sin(x),  x \epsilon [-2,2]$
2. $f(x) = e^{3x},  x \epsilon [-2,4]$    Sugerencia: usar np.e para valor de 'e'.   
3. $f(x) = \frac{1}{x},  x \epsilon [1,10]$    


</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# 1. f(x) = sin(x) en [-2, 2]
# ------------------------------
x1 = np.linspace(-2, 2, 200)
y1 = np.sin(x1)

# ------------------------------
# 2. f(x) = e^(3x) en [-2, 4]
# ------------------------------
x2 = np.linspace(-2, 4, 200)
y2 = np.exp(3 * x2)  # np.exp(x) = e^x

# ------------------------------
# 3. f(x) = 1/x en [1, 10]
# ------------------------------
x3 = np.linspace(1, 10, 200)
y3 = 1 / x3

# ------------------------------
# Graficar todas
# ------------------------------
plt.figure(figsize=(12, 8))

# f(x) = sin(x)
plt.subplot(3, 1, 1)
plt.plot(x1, y1, 'b')
plt.title("f(x) = sin(x)")
plt.grid(True)

# f(x) = e^(3x)
plt.subplot(3, 1, 2)
plt.plot(x2, y2, 'g')
plt.title("f(x) = e^(3x)")
plt.grid(True)

# f(x) = 1/x
plt.subplot(3, 1, 3)
plt.plot(x3, y3, 'r')
plt.title("f(x) = 1/x")
plt.grid(True)

plt.tight_layout()
plt.show()
```
